### Filtering Method

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
Movies = pd.read_csv('movies.csv')
Ratings = pd.read_csv('ratings.csv')
```

```python
Movies.head(-1)
```

```python
Ratings.head(-1)
```

```python
Movies.shape
```

```python
Ratings.shape
```

```python
df = pd.merge(Movies, Ratings, on = 'movieId')
```

```python
def Convert(title):
    return title.lower()
```

```python
df['title'] = df['title'].apply(Convert)
```

```python
## Calculating the mean rating of all movies
grp = df.groupby('title')['rating'].mean().sort_values(ascending = False)
```

```python
grp.head()
```

```python
dfNew = pd.DataFrame(df.groupby('title')['rating'].mean())
dfNew['NumRatings'] = df.groupby('title')['rating'].count()
```

```python
dfNew
```

```python
# User who gives rating to a movie will be displayed with columns as movie names
movieTitle = df.pivot_table(index = 'userId', columns = 'title', values = 'rating')
```

```python
movieTitle.head(20)
```

```python
cols = list(movieTitle.columns)
cols
```

```python
columns = []
for i in cols:
    columns.append(i.lower())
movieTitle.columns = columns
```

```python
movieTitle.head()
```

```python
def Recommend():
    movieName = input('Enter Movie Name').lower()
    userRatings = movieTitle[movieName]
    similar_movies = movieTitle.corrwith(userRatings)
    corr_movie = pd.DataFrame(similar_movies.dropna(), columns = ['Correlation'])
    corr_movie = corr_movie.join(dfNew['NumRatings'])
    sugg = corr_movie[corr_movie['NumRatings'] > 100].sort_values('Correlation', ascending = False)
    return sugg
```

```python
Recommend()
```

###  Content Based Approach 

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.keys()
```

```python
features = ['keywords', 'cast', 'genres','director']
```

```python
for i in features:
    df[i] = df[i].fillna("")
```

```python
def MovieCount(row):
    out = ''
    for i in features:
        out += row[i]
        out += ' '
    return out
```

```python
df['movie_content'] = df.apply(MovieCount, axis = 1)
df['movie_content']
```

```python
def LowerCaseTitle(title):
    return title.lower()
```

```python
df.title = df.title.apply(LowerCaseTitle)
```

```python
cv = CountVectorizer()
matrix = cv.fit_transform(df.movie_content)
```

```python
len(cv.get_feature_names())
```

```python
word_sim = cosine_similarity(matrix)
df['title'].head()
```

```python
def Recommendation():
    movie_name = input('Enter the Movie Name')
    movie_name = movie_name.lower().strip()
    m_index = df[df['title'] == movie_name]['index'].values[0]
    similar_word_value = word_sim[m_index]
    similar_movies = list(enumerate(similar_word_value))
    sorted_similar_movies = sorted(similar_movies, key = lambda x: x[1], reverse = True)
    movies_num = 5
    print("\nTop Similar Movies\n")
    for i in range(movies_num):
        index = sorted_similar_movies[i][0]
        movie = list(df[index:index+1]['title'])[0]
        print(i+1, movie)
```

```python
Recommendation()
```
